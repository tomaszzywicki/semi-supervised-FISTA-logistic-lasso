```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append('..')
from pathlib import Path

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from src.utils import load_arff, set_best_st_approach
from src.experiment import run_experiment, run_label_propagation_experiment
from src.visualizations import plot_experiment_results, plot_sigma_results

import warnings

warnings.filterwarnings("ignore")
```

```python
RESULTS_PATH = Path("../results").resolve()
RESULTS_PATH_LP = Path("../results/label_propagation").resolve()
```

```python
# for experiment
seeds = list(range(10))
sigma_values = [0.1, 0.3, 0.5, 1.0, 1.5, 2.0, 5.0, 10.0, 15.0, 20.0, 30.0, 40.0]
approaches = ["naive", "self_training", "label_propagation"]

# for visualizations
metrics_to_plot = ["Accuracy", "Balanced_Acc", "F1", "ROC_AUC"]

k_best = [1]

mar1_w = [1, 3]
mar1_b = [0, 2]
mar2_w = [1, 3]
mar2_b = [0, 2]
mnar_wx = [1, 3]
mnar_wy = [1, 3]
mnar_b = [0, 2]
```

```python
lsvt = load_arff("../data/raw/lsvt.arff")

X = lsvt.iloc[:, :-1].to_numpy()
y = lsvt.iloc[:, [-1]].to_numpy().ravel()
y
y = np.where(y == "2", 1, 0).astype(int)

results_lsvt = run_experiment(
    X,
    y,
    mar1_w=mar1_w,
    mar1_b=mar1_b,
    mar2_w=mar2_w,
    mar2_b=mar2_b,
    mnar_wx=mnar_wx,
    mnar_wy=mnar_wy,
    mnar_b=mnar_b,
    seeds=seeds,
    approaches=approaches,
    k_best=k_best,
    verbose=False,
    save_path=RESULTS_PATH / 'lsvt.csv'
)
```

```python
darwin = pd.read_csv("../data/raw/DARWIN.csv")

X = darwin.iloc[:, 1:-1].to_numpy()
y = darwin.iloc[:, -1].to_numpy()
y = np.where(y == "P", 1, 0)

results_darwin = run_experiment(
    X,
    y,
    mar1_w=mar1_w,
    mar1_b=mar1_b,
    mar2_w=mar2_w,
    mar2_b=mar2_b,
    mnar_wx=mnar_wx,
    mnar_wy=mnar_wy,
    mnar_b=mnar_b,
    seeds=seeds,
    approaches=approaches,
    k_best=k_best,
    verbose=False,
    save_path=RESULTS_PATH / 'darwin.csv'
)
```
